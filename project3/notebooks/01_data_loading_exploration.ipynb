```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Users/bezatezera/Desktop/Data/amazonSales/Amazon_sales_analytics/data/processed/cleaned_reviews.csv")
df.head()
```

```python
df.info()

df.describe()

df.isnull().sum()
```

```python
df['review_length'] = df['review_content'].apply(lambda x: len(x.split()))

df['review_length'].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['review_length'], bins = 30, kde=True)
plt.title("Distribution of Review Lengths")
plt.xlabel("Number of words")
plt.ylabel("Number of reviews")
plt.savefig("../results/visualization/Distribution_of_review_lengths.png")
plt.show()
```

```python
top_products = df['product_id'].value_counts().head(10)
sns.barplot(x=top_products.index, y=top_products.values)
plt.title("Top 10 Products by Number of Reviews")
plt.xlabel("Product ID")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=45)
plt.savefig("../results/visualization/top_10_products_by_reviews.png")
plt.show()
```
